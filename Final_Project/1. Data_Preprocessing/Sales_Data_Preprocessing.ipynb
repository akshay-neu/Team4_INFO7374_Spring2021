```python
StartDate_Year = []
StartDate_Month = []

for x in user_df['Subscription_StartDate'] :
    StartDate_Year.append(x.year)
    StartDate_Month.append(x.month)
```

```python
EndDate_Year = []
EndDate_Month = []

for x in user_df['Subscription_EndDate'] :
    EndDate_Year.append(x.year)
    EndDate_Month.append(x.month)
```

```python
user_df['EndDate_Year'] = EndDate_Year
user_df['EndDate_Month'] = EndDate_Month
```

```python
user_df['Total_Months_Subscription'] = (user_df['EndDate_Year'] - user_df['StartDate_Year']) * 12 + (user_df['EndDate_Month'] - user_df['StartDate_Month'])
```

```python
user_df['Subscription_Cost'] = user_df['Total_Months_Subscription'] * 10
```
