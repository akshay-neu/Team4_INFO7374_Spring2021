```python
data["Subscription_StartDate"] = pd.to_datetime(data['Subscription_StartDate'], utc=False)
```

```python
uk_data_group = data.groupby('CustomerID').agg({'Subscription_StartDate': lambda date: (date.max() - date.min()).days,
                                        'InvoiceNo': lambda num: len(num),
                                        'Subscription_Cost': lambda price: price.sum()})
```

```python
uk_data_group
```

```python

# Change the name of columns
uk_data_group.columns=['num_days','num_transactions','spent_money']
uk_data_group.head()
```

```python
# Average Order Value
uk_data_group['avg_order_value'] = uk_data_group['spent_money']/uk_data_group['num_transactions']
```

```python
uk_data_group.head()
```

```python
purchase_frequency = sum(uk_data_group['num_transactions'])/uk_data_group.shape[0]
```

```python
# Repeat Rate
repeat_rate=uk_data_group[uk_data_group.num_transactions > 74].shape[0]/uk_data_group.shape[0]
```

```python
uk_data_group[uk_data_group.num_transactions > 75].shape[0]
```

```python
repeat_rate
```

```python

#Churn Rate
churn_rate=1-repeat_rate
```

```python
purchase_frequency,repeat_rate,churn_rate
```

```python
uk_data_group['profit_margin']=uk_data_group['spent_money']*0.05
```

```python

# Customer Value
uk_data_group['CLV']=(uk_data_group['avg_order_value']*purchase_frequency)/churn_rate
```

```python
#Customer Lifetime Value
uk_data_group['cust_lifetime_value']=uk_data_group['CLV']*uk_data_group['profit_margin']
```

```python
data['month_yr'] = data['Subscription_StartDate'].apply(lambda x: x.strftime('%b-%Y'))
```

```python
data
```

```python
sale = data.pivot_table(index=['CustomerID'],columns=['month_yr'],values='Subscription_Cost',aggfunc='sum',fill_value=0).reset_index()
```

```python
sale
```

```python
# sum all months sale
sale['CLV']=sale.iloc[:,2:].sum(axis=1)
```

```python
X=sale[['Dec-2019','Nov-2019', 'Oct-2019','Sep-2019','Aug-2019','Jul-2019','Jun-2019', 'May-2019', 'Apr-2019']]
y=sale[['CLV']]
```

```python

#split training set and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=0)
```

```python
# import model
from sklearn.linear_model import LinearRegression

# instantiate
linreg = LinearRegression()

# fit the model to the training data (learn the coefficients)
linreg.fit(X_train, y_train)

# make predictions on the testing set
y_pred = linreg.predict(X_test)
```

```python

# print the intercept and coefficients
print(linreg.intercept_)
print(linreg.coef_)
```

```python
from sklearn import metrics

# compute the R Square for model
print("R-Square:",metrics.r2_score(y_test, y_pred))
```

```python
# calculate MAE using scikit-learn
print("MAE:",metrics.mean_absolute_error(y_test,y_pred))

#calculate mean squared error
print("MSE",metrics.mean_squared_error(y_test, y_pred))
# compute the RMSE of our predictions
print("RMSE:",np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
